```python
import numpy as np 
import pandas as pd 
inputdata = pd.read_csv('NYC_Crime_Cleaned.csv')
```

```python
target = inputdata["Offence_Type"].unique()
```

```python
data_dict = {}
count = 1
for data in target:
    data_dict[data] = count
    count = count + 1
inputdata["Offence_Type"] = inputdata["Offence_Type"].replace(data_dict)
```

```python
inputdata['Start_Date'] = pd.to_datetime(inputdata['Start_Date'])
```

```python
inputdata['day_of_week'] = inputdata['Start_Date'].dt.day_name()
```

```python
inputdata['day_of_week'].unique()
```

```python
district = inputdata["Offence_Level"].unique()
data_dict_district = {}
count = 1
for data in district:
    data_dict_district[data] = count
    count+=1
inputdata["Offence_Level"] = inputdata["Offence_Level"].replace(data_dict_district).dropna()
```

```python
inputdata
```

```python
inputdata = inputdata[['day_of_week','Latitude','Longitude','Offence_Level','Neighborhood','Borough']]
```

```python
days = {'Monday':0,'Tuesday':1,'Wednesday':2,'Thursday':3,'Friday':4,'Saturday':5,'Sunday':6}
inputdata['day_of_week'] = inputdata['day_of_week'].apply(lambda x: days[x])
```

```python
inputdata = inputdata.dropna()
```

```python
cat = ['day_of_week','Borough']
dummie = pd.get_dummies(inputdata[cat])
```

```python
inputdata = pd.concat([dummie,inputdata[['Latitude','Longitude','Offence_Level','Neighborhood']]],axis = 1)
```

```python
from sklearn import linear_model, neighbors, preprocessing
min_max_scaler = preprocessing.MinMaxScaler()
dummie_maxmin = min_max_scaler.fit_transform(inputdata.drop('Offence_Level',axis=1))
```

```python
from sklearn.model_selection import RandomizedSearchCV, GridSearchCV, train_test_split, cross_val_score
X_train,X_test, y_train, y_test = train_test_split(inputdata.drop('Offence_Level',axis=1),inputdata['Offence_Level'],test_size=0.3,random_state=1)
```

```python
X_train = X_train.values
X_test = X_test.values 
y_train = y_train.values
y_test = y_test.values
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 30)
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)
predictions
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,predictions)
```

```python
cm = confusion_matrix(predictions,y_test)
cm
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100,random_state=0,criterion='gini',class_weight='balanced')
clf.fit(X_train,y_train)
y_predict = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test,y_predict)
```
